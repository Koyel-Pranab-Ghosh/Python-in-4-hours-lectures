## <font color="red">for loop</font>

## You can use for loop in two ways. 

## <font color="green">Method 1 syntax
## for i in range(start, stop, step_size):
## .... Do something here </font>

## <font color="blue">Method 2 syntax
## for i in List_Or_Tuple_Name:
## .... Do something here</font>


```python
## Method_1
for i in range(1,10,2):
    print(i)
```

```python
## Method_2
My_List_06 = [23,12,89,47,62,50]
for i in My_List_06:
    print(i)
```

## Instead of writing 'i' above, we could have named it as anything ..cat,dog,owl ..etc.. but whatever you name it in for loop remeber to use the same name inside. Also notice that after for there is TAB for the next line onwards

```python
## Method_1
for abra_ca_dabra in range(1,10,2):
    print(abra_ca_dabra)
```

```python
## Method_2
My_List_06 = [23,12,89,47,62,50]
for rum_pum_rum in My_List_06:
    print(rum_pum_rum)
```

## Below we will see how to make use of it. Suppose we want to know, square and cube of every number from 90,100 we will do it as below using method_1

```python
for i in range(90,101,1):
    print(i**2, " ", i**3)
```

## <font color="red"> while loop </font>

## in while loop, you test a condition. Until that test conditions holds true you repeat the task always. Two things
## <font color="red"> Most of the time, you have to start that test condition </font>
## <font color="red"> Dont get caught in infinite loop otherwise the task will never be over </font>

## It is a good idea at this moment to locate the stop button on top of this page (The black filled square button). If you ever need to stop a program just press that.

## Let us see how to use while loop to achieve the same task that we did with for loop

```python
Current_Number = 90
while(Current_Number <= 100):
    print(Current_Number**2, Current_Number**3)
    Current_Number += 1 ## This is the most important line in while. You need to increase or decrease the test value
```

## As you can see we achieved the same output as for loop

<ol>
    <h2>
        <li> First line in the above program initiatlizes the test condition </li>
        <li> Second line starts the while loop and checks if the test condition is passed </li>
        <li> Third line does the computation .. In practise it can be many many lines </li>
        <li> The last line changes the test condition. In this case we increased by 1 but we could have increased or decreased by any value we wish. <font color="red"> Please note that this is the most important line if you want the while loop to stop and behave correctly</font></li>
        <li> Please note that from 3rd line onwards the code is shifted to right by a TAB.. This shows that line 3 and 4 are part of while loop
    </h2>
</ol>

##  Now consider this.

## Let us break the while loop intentionally. Make sure you have located the Stop button in this page. You will need it.

## Before you run the program know what is going to happen below

## While loop starts but the 4th line is shifted to LEFT by a TAB. That means that 4th line is no longer a part of the while loop. That in turn means that Current_Number never increases and is always stuck at 90 forever...it will keep printing the same value again and again

## <font color="red"> Watch out for these kind of mistakes. It is very easy to make these mistakes. </font>


```python
Current_Number = 90
while(Current_Number <= 100):
    print(Current_Number**2, Current_Number**3)
Current_Number += 1 ## We have made a mistake. We took this line out of reach of while
```

# Using for loop go through each number from 101 to 399 in steps of 11 and
## print "number is divisible by 2" if the number % 2 is equal to 0
## print "number is divisible by 3" if the number % 3 is equal to 0
## print "number is divisible by 5" if the number % 5 is equal to 0
## print "number is divisible by 7" if the number % 7 is equal to 0

```python
for i in range(101,400,11):
    if i%2 == 0:
        print(i, " is divisible by 2")
    if i%3 == 0:
        print(i, " is divisible by 3")
    if i%5 == 0:
        print(i, " is divisible by 5")
    if i%7 == 0:
        print(i, " is divisible by 2")
```

# Using while loop go through each number from 101 to 399 in steps of 11 and
## print "number is divisible by 2" if the number % 2 is equal to 0
## print "number is divisible by 3" if the number % 3 is equal to 0
## print "number is divisible by 5" if the number % 5 is equal to 0
## print "number is divisible by 7" if the number % 7 is equal to 0

```python
Current = 101
while(Current <= 399):
    if Current%2 == 0:
        print(Current, " is divisible by 2")
    if Current%3 == 0:
        print(Current, " is divisible by 3")
    if Current%5 == 0:
        print(Current, " is divisible by 5")
    if Current%7 == 0:
        print(Current, " is divisible by 2")   
    Current += 11 ## This is most important line if you want the while loop to end. Otherwise it will go on for ever
```
