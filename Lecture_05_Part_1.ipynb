## Now that we know how to create an empty figure, let us add something to it.

## Suppose we have a list of 10 consecutive numbers and we want 1 subplot inside the bigger plot. In it we want the list to be on x-axis and the X raised to the power 2 on the Y axis. 


## Notice carefully how we use it. 

## Steps 

<ol>
    <li> Prepare X-axis data </li>
    <li> Prepare Y-axis data </li>
    <li> create fig and axes_list for a subplots with nrows=1 and ncols=1 </li>
    <li> plot X and Y axis data on the axes_list </li>
    <li> Finally show the figure </li>
</ol>

```python
X_Axis_data = list(range(0,10,1)) 
Y_Axis_data = [] 
for each_X_point in X_Axis_data:
    Y_Axis_data.append(each_X_point**2)
```

```python
fig, axes_list = plt.subplots(nrows=1, ncols=1) ## Step 3
axes_list.plot(X_Axis_data, Y_Axis_data) ## Step 4
## Showing step in this case will be done automatically
```

## Ok nice. Let us change the color of the curve from blue to red. Notice subtle change.

```python
fig, axes_list = plt.subplots(nrows=1, ncols=1) ## Step 3
plt.plot(X_Axis_data, Y_Axis_data, color="red") ## Step 4
## Showing step in this case will be done automatically
```

## OK. We also want to see individual data points apart from the continuous line that connects it. Note exactly at which place we are adding instruction compared to above. We need to make two plots ..one for the individual data points and the other for the line. In a single figure you can keep adding as many plots as you need.

```python
fig, axes_list = plt.subplots(nrows=1, ncols=1) ## Step 3
plt.plot(X_Axis_data, Y_Axis_data, "o", color="blue") ## Step 4
plt.plot(X_Axis_data, Y_Axis_data, color="red") ## Step 4
```

## There is no description to the X-axis .. Let us put the description ..."This is X-axis of my data". This description of X-axis is called ''xlabel". See how we use it.

```python
fig, axes_list = plt.subplots(nrows=1, ncols=1) 
plt.plot(X_Axis_data, Y_Axis_data, "o", color="blue") 
plt.plot(X_Axis_data, Y_Axis_data, color="red") 
plt.xlabel("This is X-axis of my data")
```

```python
## Similarly we can add Y-axis label, called 'ylabel' , to the curve
fig, axes_list = plt.subplots(nrows=1, ncols=1) 
plt.plot(X_Axis_data, Y_Axis_data, "o", color="blue") 
plt.plot(X_Axis_data, Y_Axis_data, color="red") 
plt.xlabel("This is X-axis of my data")
plt.ylabel("Square of X-axis")
```

## Ok good but there is no main title of the graph till now. We can set the title using 'title' function of pyplot. Notice the usage.

```python
fig, axes_list = plt.subplots(nrows=1, ncols=1) 
plt.plot(X_Axis_data, Y_Axis_data, "o", color="blue") 
plt.plot(X_Axis_data, Y_Axis_data, color="red") 
plt.xlabel("This is X-axis of my data")
plt.ylabel("Square of X-axis")
plt.title("Plot of a number and its square for numbers from 0 to 9")
```

## Let us see another example.

## Suppose we have a list of 50 values , ranging from 0 to 2*pi and each point increases by pi/25.
## Also suppose we have to make two subplots (side by side). In the left we have to plot sin of all values while on the right we have to plot cos of all values. Pause and read through line by line below to ensure you understand every command. Make sure you read the comments

```python
import math # you need this to do math operations like sin, cos, pi etc.

## Create X axis data
X_axis_data = []
StartVal = 0
while(StartVal <= 2*math.pi):
    X_axis_data.append(StartVal)
    StartVal += math.pi/25
## Create Y axis data for left figure which is sin of x axis
Y_axis_data_Sin = []
for each_X in X_axis_data:
    Y_axis_data_Sin.append(math.sin(each_X))
## Create Y axis data for right figure which is cos of x axis
Y_axis_data_Cos = []
for each_X in X_axis_data:
    Y_axis_data_Cos.append(math.cos(each_X))
    
## Create a subplots of 1 row and 2 columns. 
fig, axes_list = plt.subplots(nrows=1, ncols=2)

## Now there are two plots. Each has roll number like (totalrows, totalcols, figurenumber).
## Thus the subplots are (1,2,1) and (1,2,2). 
## Please make sure you understand the above because we need this roll number to switch to a figure in multi-plot
## setting

## Let us make the sin graph
plt.subplot(1,2,1) ## Notice the roll number
plt.plot(X_axis_data, Y_axis_data_Sin, color="red")
plt.xlabel("Angle in pi")
plt.ylabel("Sin of angle")
plt.title("Sin of angle in the range 0 to 2*pi insteps of pi/25")

## Let us make the cos graph
plt.subplot(1,2,2) ## Notice the roll number
plt.plot(X_axis_data, Y_axis_data_Cos, color="blue")
plt.xlabel("Angle in pi")
plt.ylabel("Cos of angle")
plt.title("Cos of angle in the range 0 to 2*pi insteps of pi/25")
```

## Good! We got the plot but we have a problem. Since the plots are smal the title are going into each other.
## Let us fix it.
## You can adjust the figure size by 'figsize=(x,y)'.
## Notice the place and use of the statement

```python
## Create a subplots of 1 row and 2 columns. 
fig, axes_list = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

## Now there are two plots. Each has roll number like (totalrows, totalcols, figurenumber).
## Thus the subplots are (1,2,1) and (1,2,2). 
## Please make sure you understand the above because we need this roll number to switch to a figure in multi-plot
## setting

## Let us make the sin graph
plt.subplot(1,2,1) ## Notice the roll number to switch to this subplot
plt.plot(X_axis_data, Y_axis_data_Sin, color="red")
plt.xlabel("Angle in pi")
plt.ylabel("Sin of angle")
plt.title("Sin of angle in the range 0 to 2*pi insteps of pi/25")

## Let us make the cos graph
plt.subplot(1,2,2) ## Notice the roll number to switch to this subplot
plt.plot(X_axis_data, Y_axis_data_Cos, color="blue")
plt.xlabel("Angle in pi")
plt.ylabel("Cos of angle")
plt.title("Cos of angle in the range 0 to 2*pi insteps of pi/25")
```

## Now suppose we have more than one plot in a single subplot, then we need something called legend. Legend is nothing but a description of different curves in a single page.

## Suppose we have twelve months data for the previous year and we have temperature, humidity, windspeed, rainfall (in mm) and pressure recorded for each month.

## Also suppose, for the purpose of demonstration only, that we have to plot all the five properties of each month in a single plot. 

## See below how we adjust. Let us first create the dataset


```python
import random
random.seed(123456)
## First create the random data
Month_X_Axis = list(range(1,13,1))
## Tempearture
Temperature = []
for month in Month_X_Axis:
    Temperature.append(random.randint(10,50))
## humidity
Humidity = []
for month in Month_X_Axis:
    Humidity.append(random.randint(12,60))
## windspeed
Windspeed = []
for month in Month_X_Axis:
    Windspeed.append(random.randint(8,30))    
## windspeed
Rainfall = []
for month in Month_X_Axis:
    Rainfall.append(random.randint(20,70)) 
## pressure
pressure = []
for month in Month_X_Axis:
    pressure.append(random.randint(5,15))     
```

## Now that we have the data, let us go ahead and plot these in a single plot. We will need to use plt.legend. For that to be used you have to save each curve to a name. After that note how we have used it inside plt.legend()

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
Plot_1 = plt.plot(Month_X_Axis, Temperature, color="red")
Plot_2 = plt.plot(Month_X_Axis, Humidity, color="green")
Plot_3 = plt.plot(Month_X_Axis, Windspeed, color="blue")
Plot_4 = plt.plot(Month_X_Axis, Rainfall, color="brown")
Plot_5 = plt.plot(Month_X_Axis, pressure, color="black")
plt.xlabel("Month number")
plt.ylabel("Various values")
plt.title("Various paraeteres of every month")
plt.legend((Plot_1[0], Plot_2[0], Plot_3[0], Plot_4[0], Plot_5[0]),
           ("Temperature", "Humidity","Windspeed","Rainfall","pressure"))
```

## We will create a subplot with 6 rows and 1 columns. We will se 6 different types of charts.


```python
fig, axes_list = plt.subplots(nrows=6, ncols=1, figsize=(5,25))
## Line graph as a reference which we have been drawing till now
plt.subplot(6,1,1)
plt.plot(Month_X_Axis, Temperature)
plt.title("Line graph")

## Scatter plot
plt.subplot(6,1,2)
plt.scatter(Humidity, Temperature)
plt.title("Scatter graph")

## Histogram plot
plt.subplot(6,1,3)
plt.hist(Temperature)
plt.title("Histogram graph")

## Pie chart
plt.subplot(6,1,4)
plt.pie(Temperature)
plt.title("pie chart")

## Bar chart
plt.subplot(6,1,5)
plt.bar(Month_X_Axis, Temperature)
plt.title("bar chart")

## Heat map
plt.subplot(6,1,6)
plt.imshow([Humidity, Temperature])
plt.title("Heat Map")
```
